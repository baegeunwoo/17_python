```python
# else와 들여쓰기
x = 5


if x == 10:
    print("10 입니다.")
else:
    print("x에 들어있는 숫자는 ")
    print("10이 아닙니다")
```

```python
# if 조건문에 숫자 지정하기

if 0:
    print("참")
else:
    print("거짓")

if 1:
    print("참")
else:
    print("거짓")

if 0x1f:
    print("참")
else:
    print("거짓")

if 0b1000:
    print("참")
else:
    print("거짓")

if 13.5:
    print("참")
else:
    print("거짓")

```

```python
# 조건문에 문자열 지정하기
if "hello":
    print("참")
else:
    print("거짓")

if " ":
    print("참")
else:
    print("거짓")

    
if "":
    print("참")
else:
    print("거짓")
```

```python
if not 0:
    print("참")

if not None:
    print("참")

if not "":
    print("참")
```

```python
# 조건식을 여러개 지정하기

x = 10
y = 20

if x == 10 and y == 20:
    print("참")
else:
    print("거짓")
```

```python
x = 15

if x > 0 and x < 20:
    print("입장 가능합니다.")

if 0 < x < 20:
    print("입장 가능합니다.")

if x > 0:
    if x < 20:
        print("입장 가능합니다.")


y=70
if y<8 or y >= 65:
    print("할인 대상입니다. ")
```
