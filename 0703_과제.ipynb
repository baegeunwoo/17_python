```python
students = {

    "김철수": [90, 80, 70],
    
    "이영희": [100, 95, 90],
    
    "박민수": [60, 70, 65]

}
```

```python
print("===== 학생 성적 결과 =====\n")

total=0
for key,value in students.items():
    print(f"이름:{key}\n")
    print(f"점수:{value}\n")
    print(f"평균:{sum(value)/len(students):.1f}\n\n")

    total += sum(value)

print("===== 전체 결과 =====")

print(f"전체 학생 수:{len(students)}\n")
print(f"전체 평균:{total/(len(students)*3)}\n")
```
